##############SELECTION SORT ######################
Write a program to sort an array of N integers in ascending order using the Selection Sort algorithm without using built-in sorting functions.

```python
n=int(input())
arr=list(map(int,input().split()) )
for i in range(n):
  min_index=i
  for j in range(i+1,n):
    if(arr[j]<=arr[min_index]):
      min_index=j
  arr[i],arr[min_index]=arr[min_index],arr[i]
print(*arr)
```
