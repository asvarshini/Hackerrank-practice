You are given a sorted array numbers and an integer target.
Find two numbers in the array such that their sum equals the target.

Return the 1-indexed positions of the two numbers.

Conditions

numbers[i] + numbers[j] = target

1 ≤ i < j ≤ n

The same element cannot be used twice.

Exactly one solution exists.

```python
n = int(input())
arr = list(map(int, input().split()))
target = int(input())

l = 0
r = n - 1

while l < r:
    s = arr[l] + arr[r]

    if s == target:
        print(l+1, r+1)
        break
    elif s < target:
        l += 1
    else:
        r -= 1
```
